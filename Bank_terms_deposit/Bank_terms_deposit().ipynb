```python
df['Deposit'].value_counts()
```

This target feature is highly imbalanced.
to resolve this issues i will use  Oversampling the Minority Class ,Synthetic Minority Oversampling Technique. (SMOTE)

Labeling the Features.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
for columns in categorical_df.columns:
  df[columns]=le.fit_transform(df[columns])
```

applying the SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote=SMOTE()
```

```python
X_resampled,y_resampled=smote.fit_resample(df.drop('Deposit',axis=1),df['Deposit'])
```

```python
y_resampled.value_counts()
```

```python
sns.countplot(data=df, x=y_resampled,color='red')
plt.xticks(rotation=90)
plt.show()
```

Scaling the features.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_resampled=scaler.fit_transform(X_resampled)
#scaler.transform(X_resampled)

```

Dividing the data into train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X_resampled_train,X_resampled_test,y_resampled_train,y_resampled_test=train_test_split(X_resampled,y_resampled,test_size=0.3,random_state=42,stratify=y_resampled)
```

```python
# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200],  # Number of trees
    'learning_rate': [0.01, 0.1, 0.2],  # Step size shrinkage
    'max_depth': [3, 5, 7],  # Maximum depth of trees
    'min_child_weight': [1, 3],  # Minimum sum of instance weight
    'subsample': [0.8, 1],  # Fraction of samples for training
    'colsample_bytree': [0.8, 1]  # Fraction of features for each tree
}
```

```python
grid_search = GridSearchCV(xgb_model, param_grid, cv=3, scoring='accuracy',)
```

```python
grid_search.fit(X_resampled_train, y_resampled_train)
```

```python
# Best parameters and model
best_xgb = grid_search.best_estimator_
print(f"Best Parameters: {grid_search.best_params_}")
```
