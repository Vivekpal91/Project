```python
#step 2 define the paragraph (context)

paragraph = '''
Artificial Intelligence (AI) is a branch of computer science focused on creating systems capable of performing tasks that typically require human intelligence.
These tasks include problem-solving, decision-making, learning, language understanding, and perception.
AI is broadly categorized into narrow AI, which is designed for specific tasks like facial recognition or language translation, and general AI,
which aims to replicate human cognitive abilities across a wide range of activities. With advancements in machine learning, deep learning, and natural language processing,
AI has become integral to industries such as healthcare, finance, transportation, and entertainment.
Its potential to revolutionize fields like personalized medicine, autonomous vehicles, and intelligent automation continues to drive innovation and ethical debates worldwide.'''

```

```python
#step 3 :define the list of question to ask

questions = [
    'What is Artificial Intelligence, and what tasks is it designed to perform?',
    'How is AI categorized, and what are the differences between narrow AI and general AI?',
    'What are some examples of tasks that AI can perform?',
     'What are the key technologies or methodologies driving AI advancements, such as machine learning and deep learning?',
    'How does natural language processing contribute to AI applications?'
]
```

```python
# step 4 : ask question and get answers from the paragraph

for question in questions:
  result = qa_pipeline(question=question,context=paragraph)
  answer = result['answer']
  print(f"Q: {question}")
  print(f"A: {result['answer']}\n")
```

```python
#step 2: function to extract text froam a pdf file
```

```python
def extract_text_from_pdf(Pdf_path):
  pdf_text = ''
  with open(pdf_path, 'rb') as file:
    reader = PyPDF2.PdfReader(file)
    for page in reader.pages:
      pdf_text = pdf_text + page.extract_text() + '\n'
  return pdf_text
```
