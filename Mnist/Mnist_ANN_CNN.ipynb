```python
y_cat_train=to_categorical(y_train)
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train
```

```python
X_train[2]
```

```python
 X_train=X_train.reshape(60000,28,28,1) # value 3 is for colorfull image
```

```python
X_test = X_test.reshape(10000,28,28,1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model=Sequential()
```

```python
model.add(Conv2D(filters=32,kernel_size=(4,4),input_shape=(28,28,1),activation='relu'))
```

```python
model.add(MaxPool2D(pool_size=(2,2))) #model.add(MaxPool2D(pool_size=(2,2),strides=1))
```

```python
model.add(Conv2D(filters=16,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Conv2D(filters=14,kernel_size=(3,3),activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
# ANN
```

```python
model.add((Dense(128,activation='relu'))) # hidden layer
model.add((Dense(64,activation='relu')))
```

```python
model.add((Dense(10,activation='softmax'))) # output layer
```

```python
model.compile(loss='categorical_crossentropy')
```

```python
model.fit(X_train,y_cat_train,epochs=2)
```

```python
model.evaluate(X_test,y_cat_test)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_pred = 1 if y_pred > 0.5 else 0
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(f'{classification_report(y_test,y_pred.argmax(axis=1))}')
```
